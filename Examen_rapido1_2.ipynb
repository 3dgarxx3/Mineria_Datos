```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
# 1. Cargar los datos
df = pd.read_csv('munsell-array-fixed-choice.csv')
```

```python
# 2. Configurar la figura para 3 gráficas de barras
fig, axes = plt.subplots(3, 1, figsize=(10, 18))

idiomas = ['english_color', 'spanish_color', 'tsimane_color']
titulos = ['Distribución de Colores: Inglés', 'Distribución de Colores: Español', 'Distribución de Colores: Tsimane']

for i, idioma in enumerate(idiomas):
    # Contar la frecuencia de cada color
    counts = df[idioma].value_counts()

    # Crear la gráfica de barras
    counts.plot(kind='bar', ax=axes[i], color='skyblue', edgecolor='black')

    axes[i].set_title(titulos[i], fontsize=14)
    axes[i].set_ylabel('Frecuencia (Cantidad de celdas)')
    axes[i].set_xlabel('Categoría de Color')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
