IPython Magic Commands

```python
# Sirve para evaluar el desempeño de una expreson en cuanto a tiempo
%timeit L = [n ** 2 for n in range(1000)]
```

```python
# Con doble porcentaje podemos evaluar el desempeño
# en cuanto tiempo de mas de una linea
%%timeit
L = []
for n in range(1000):
  L.append(n ** 2)
```

```python
# importa numpy y renombralo como np
import numpy as np
```

```python
# Arreglo lineal de enteros
np.array([1, 4, 2, 5, 3])
```

```python
# Al ingresar enteros y flotantes los convierte en flotantes
np.array([3.14, 4, 2, 3])
```

```python
np.array([1, 2, 3, 4], dtype='float32')
```

```python
np.array([1, 2, 3, 4], dtype='int')
```

```python
# Cuando el argumento de np.array es una iteraccion anidada
#genera una matriz
np.array([range(i, i + 3) for i in [2, 4, 6]])
```
