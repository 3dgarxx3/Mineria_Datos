```python
dfc = pd.read_csv("clientes.csv")
dfo = pd.read_csv("ordenes.csv")
```

```python
df_merge = dfc.merge(dfo, on="cliente_id", how="left")
```

```python
total_filas = len(df_merge)
print("Total de filas:", total_filas)
```

```python
monto_por_cliente = (
    df_merge
    .groupby("cliente_id")["monto"]
    .sum()
    .fillna(0)
)
print(monto_por_cliente)
```

```python
df = pd.read_csv(
    "sensores.csv",
    usecols=["ts", "temperatura", "humedad"]
)
```

```python
df["ts"] = pd.to_datetime(df["ts"], format="mixed", errors="coerce")
```

```python
mediana_temp = df["temperatura"].median()
media_hum = df["humedad"].mean()

df["temperatura"] = df["temperatura"].fillna(mediana_temp)
df["humedad"] = df["humedad"].fillna(media_hum)
```

```python
nat_ts = df["ts"].isna().sum()
```

```python
mediana_temp_final = round(df["temperatura"].median(), 1)
media_hum_final = round(df["humedad"].mean(), 3)
print(nat_ts, mediana_temp_final, media_hum_final)
```
