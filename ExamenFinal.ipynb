```python
import pandas as pd
import numpy as np

# Establece una semilla para reproducibilidad
np.random.seed(42)

# 1. Crea una lista de matrículas de estudiantes
student_ids = [f'{i:06d}' for i in range(172120, 172120 + 20)]

# 2. Asigna 10 estudiantes al grupo 'Café' y 10 al grupo 'Agua'
groups = ['Café'] * 10 + ['Agua'] * 10
np.random.shuffle(groups)

# 3. Simula calificacionen de examenes para el grupo café
# Mean=70, Std=5, asegurate de que los valores estén entre (50-100)
coffee_scores = np.random.normal(loc=70, scale=5, size=10).round(0)
coffee_scores = np.clip(coffee_scores, 50, 100).astype(int)

# 4. Simula calificaciones de examen para el grupo de 'Agua'
# Mean=75, Std=5, asegurate de que los valores estén entre (50-100)
water_scores = np.random.normal(loc=75, scale=5, size=10).round(0)
water_scores = np.clip(water_scores, 50, 100).astype(int)

# Combina los valores basado en el grupo
exam_scores = []
coffee_idx = 0
water_idx = 0
for group in groups:
    if group == 'Café':
        exam_scores.append(coffee_scores[coffee_idx])
        coffee_idx += 1
    else:
        exam_scores.append(water_scores[water_idx])
        water_idx += 1

# 5. Combina en un DataFrame de Pandas
df_students = pd.DataFrame({
    'Matrícula': student_ids,
    'Grupo': groups,
    'Calificaciones': exam_scores
})

print("Datos simulados:")
print(df_students.head())


# =========================================================
# NUEVO CÓDIGO: Cálculo de promedios para Café y Agua
# =========================================================

print("\n" + "="*40)
print("PROMEDIO DE CALIFICACIONES POR GRUPO")
print("="*40)

# Método rápido usando groupby de Pandas
promedios = df_students.groupby('Grupo')['Calificaciones'].mean()
print("Resumen con groupby:")
print(promedios)

# Método detallado filtrando de forma individual
promedio_cafe = df_students[df_students['Grupo'] == 'Café']['Calificaciones'].mean()
promedio_agua = df_students[df_students['Grupo'] == 'Agua']['Calificaciones'].mean()

print(f"\nEl promedio del grupo que tomó Café es: {promedio_cafe}")
print(f"El promedio del grupo que tomó Agua es: {promedio_agua}")
```

#Tabla de contigencia


```python
import pandas as pd

# 1. Creamos un diccionario con los datos (excluyendo los totales para que no interfieran en los cálculos)
datos = {
    'Aprobó': [45, 20],
    'Reprobó': [5, 30]
}

# 2. Convertimos el diccionario en un DataFrame asignando los nombres de las filas
df_contingencia = pd.DataFrame(datos, index=['Tomó el curso', 'No tomó el curso'])

print("Tabla de Contingencia Base:")
print(df_contingencia)
```

```python
# 1. Obtener el número de estudiantes que tomaron el curso Y aprobaron
# Usamos .loc[fila, columna]
favorables = df_contingencia.loc['Tomó el curso', 'Aprobó']

# 2. Obtener el total de estudiantes que tomaron el curso (sumando la fila completa)
total_curso = df_contingencia.loc['Tomó el curso'].sum()

# 3. Calcular la probabilidad condicional
probabilidad_condicional = favorables / total_curso

# 4. Mostrar el resultado en diferentes formatos
print(f"Casos favorables (Tomó el curso y Aprobó): {favorables}")
print(f"Total del grupo (Tomó el curso): {total_curso}")
print("-"*50)
print(f"Probabilidad condicional en decimal: {probabilidad_condicional:.4f}")
print(f"Probabilidad condicional en porcentaje: {probabilidad_condicional * 100:.2f}%")
```

```python
total_aprobaron = df_contingencia['Aprobó'].sum()

# 2. Obtener el total general de estudiantes en el estudio (sumando todos los valores de la tabla)
total_general = df_contingencia.values.sum()

# 3. Calcular la probabilidad simple
probabilidad_simple = total_aprobaron / total_general

print(f"Total de estudiantes que aprobaron: {total_aprobaron}")
print(f"Total general de estudiantes: {total_general}")
print("-"*50)
print(f"Probabilidad simple en decimal: {probabilidad_simple:.4f}")
print(f"Probabilidad simple en porcentaje: {probabilidad_simple * 100:.2f}%")
```

RMSE

```python
import numpy as np
from sklearn.metrics import mean_squared_error

# 1. Definir los datos del problema
observados = np.array([3, 7, 5, 10])
predichos = np.array([2, 8, 4, 9])

# Método 1: Usando scikit-learn
# (Nota: En versiones recientes de sklearn se usa root_mean_squared_error directamente,
# o MSE con squared=False)
rmse_sklearn = mean_squared_error(observados, predichos, squared=False)

# Método 2: Cálculo manual con numpy siguiendo la fórmula
mse_manual = np.mean((observados - predichos) ** 2)
rmse_manual = np.sqrt(mse_manual)

print("=== RESULTADOS DEL CÁLCULO DE RMSE ===")
print(f"RMSE (vía scikit-learn): {rmse_sklearn}")
print(f"RMSE (cálculo manual):   {rmse_manual}")
```

#Maquina defectuosa

```python
# Definición de variables con los datos del problema
p_A = 0.60
p_B = 0.40

p_D_dado_A = 0.02
p_D_dado_B = 0.05

# 1. Ley de Probabilidad Total
p_D = (p_A * p_D_dado_A) + (p_B * p_D_dado_B)

# 2. Teorema de Bayes
p_B_dado_D = (p_B * p_D_dado_B) / p_D

print(f"Probabilidad total de pieza defectuosa: {p_D:.3f} ({p_D*100:.1f}%)")
print(f"Probabilidad de que sea de la Máquina B dado que es defectuosa: {p_B_dado_D:.4f} ({p_B_dado_D*100:.2f}%)")
```

```python
import pandas as pd

# Cargar archivos (ajusta la ruta si los subes manualmente)
dfc = pd.read_csv('/content/clientes.csv')
dfo = pd.read_csv('/content/ordenes.csv')

# Left merge
df_merge = pd.merge(dfc, dfo, on='cliente_id', how='left')

# (a) Número total de filas
num_filas = len(df_merge)
print("Número total de filas:", num_filas)

# (b) Monto total por cliente (NaN -> 0)
monto_por_cliente = (
    df_merge
    .groupby('cliente_id')['monto']
    .sum(min_count=1)  # evita que todo NaN se vuelva 0 de inmediato
    .fillna(0)         # reemplaza NaN por 0
    .reset_index()
)

print("\nMonto total por cliente:")
print(monto_por_cliente)

# Valor específico solicitado
monto_cliente_1 = monto_por_cliente.loc[
    monto_por_cliente['cliente_id'] == 1, 'monto'
].values[0]

print("\nMonto total cliente_id=1:", monto_cliente_1)
```
